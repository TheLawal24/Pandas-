````python
# @title Unnamed: 0 vs Maximum Takeoff Weight (kg)

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Unnamed: 0']
  ys = series['Maximum Takeoff Weight (kg)']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = aircraft_df.sort_values('Unnamed: 0', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Engine Type')):
  _plot_series(series, series_name, i)
  fig.legend(title='Engine Type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Unnamed: 0')
_ = plt.ylabel('Maximum Takeoff Weight (kg)')
````

````python
# @title Unnamed: 0 vs Empty Weight (kg)

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Unnamed: 0']
  ys = series['Empty Weight (kg)']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = aircraft_df.sort_values('Unnamed: 0', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Center of Gravity Limits')):
  _plot_series(series, series_name, i)
  fig.legend(title='Center of Gravity Limits', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Unnamed: 0')
_ = plt.ylabel('Empty Weight (kg)')
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X = aircraft_df.drop(['Unnamed: 0', 'Aircraft Model','Center of Gravity Limits', 'Engine Type'], axis=1)
y = aircraft_df['Aircraft Model']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

n_neighbors = 20

clf = KNeighborsClassifier(n_neighbors=n_neighbors)

clf.fit(X_train, y_train)
````

````python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the model
model = RandomForestClassifier()
````

````python
y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
````
